```python
from manim import *

class BulletedListExample(Scene):
    def construct(self):
        items = [
            Text("العنصر 1"),
            Text("العنصر 2"),
            Text("العنصر 3")
        ]
        
        blist = VGroup(*items)
        for i, item in enumerate(blist):
            item.next_to(LEFT * 0.5, DOWN * 0.5 + i * DOWN * 0.5)
            dot = MathTex("\\cdot").scale(0.5).next_to(item, LEFT)
            self.add(dot, item)
            if i == 0:
                item.set_color(RED)
            elif i == 1:
                item.set_color(GREEN)
            elif i == 2:
                item.set_color(BLUE)

# Don't remove the command below for the run button to work
%manim -ql -v WARNING BulletedListExample

```

```python
from manim import *

class BulletedListExample(Scene):
    def construct(self):
        items = ["العنصر 1", "العنصر 2", "العنصر 3", "العنصر 4", "العنصر 5"]  # Add more items as needed
        vertical_distance = 0.7  # Adjust this value to change vertical distance
        dot_size = 0.5  # Adjust this value to change dot size
        color_list = [RED, GREEN, BLUE, YELLOW, PURPLE]  # Add more colors as needed
        horizontal_offset = 1.5  # Adjust this value to change horizontal offset
        vertical_offset = -3  # Adjust this value to change vertical offset
        
        blist = VGroup()
        for i, item_text in enumerate(items):
            item = Text(item_text)
            item.next_to(LEFT * horizontal_offset, DOWN * (vertical_offset + vertical_distance) + i * DOWN * vertical_distance)
            dot = MathTex("\\cdot").scale(dot_size).next_to(item, LEFT)
            self.add(dot, item)
            blist.add(dot, item)
            
            # Color the items based on cycling through the color_list
            item.set_color(color_list[i % len(color_list)])

# Don't remove the command below for the run button to work
%manim -ql -v WARNING BulletedListExample
```
